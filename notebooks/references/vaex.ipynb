# Visualizing the NYC Taxi dataset

```python
long_min = -74.05
long_max = -73.75
lat_min = 40.58
lat_max = 40.90
```

```python
taxi1.viz.heatmap(taxi1.pickup_longitude, taxi1.pickup_latitude, f="log1p", limits=[[long_min, long_max], [lat_min, lat_max]], show=True);
```
